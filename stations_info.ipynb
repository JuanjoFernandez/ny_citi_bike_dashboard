```python
# Libraries and dependencies
import pandas as pd
import json, requests
import os
```

```python
# Retrieving the bike stations info JSON
url = 'https://gbfs.citibikenyc.com/gbfs/en/station_information.json'
response = requests.get(url)
station_json = response.json()

# Saving to a dataframe
station_df = pd.json_normalize(station_json["data"]["stations"])

station_df
```

```python
# Dropping unecessary columns

station_df.drop(columns = ['station_type', 'region_id', 'short_name', 'rental_methods', 'legacy_id', 'has_kiosk',
                         'eightd_has_key_dispenser', 'external_id', 'eightd_station_services', 'electric_bike_surcharge_waiver',
                           'rental_uris.android', 'rental_uris.ios'], inplace = True)
```

```python
# Rearranging and indexing
station_df = station_df[['station_id', 'name', 'lat', 'lon', 'capacity']]
station_df.set_index('station_id', drop=True, inplace=True)
station_df
```

```python
# Saving the clean data to a csv
path = os.path.join('resources', 'clean_station_info.csv')
station_df.to_csv(path)
```
