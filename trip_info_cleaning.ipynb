```python
# Libraries and dependencies
import pandas as pd
import os
```

```python
# Opening the csv files

# Saving a list for all files in the directory
files = [f for f in os.listdir("./resources/historical") if f.endswith('.csv')]

# Creating the filepath list
filespath = []
for f in files:
    filespath.append(os.path.join ('resources', 'historical', f))

# Creating the raw dataframe
csv_files = []
for file in filespath:
    csv_files.append(pd.read_csv(file, index_col = None, header = 0))

trip_df = pd.concat(csv_files, axis=0, ignore_index=True)

trip_df
```

```python
# Cleaning the dataframe, mostly because of the column name changes, will keep the column names consistent
# with the most current file
# Column list:
# ride_id,rideable_type,started_at,ended_at,start_station_name,start_station_id,end_station_name,end_station_id,
# start_lat,start_lng,end_lat,end_lng,member_casual

clean_trip_df = trip_df

# Merging columns
clean_trip_df['started_at']=trip_df['started_at'].combine_first(trip_df['starttime'])
clean_trip_df['ended_at']=trip_df['ended_at'].combine_first(trip_df['stoptime'])
clean_trip_df['start_station_name']=trip_df['start_station_name'].combine_first(trip_df['start station name'])
clean_trip_df['end_station_name']=trip_df['end_station_name'].combine_first(trip_df['end station name'])
clean_trip_df['start_station_id']=trip_df['start_station_id'].combine_first(trip_df['start station id'])
clean_trip_df['end_station_name']=trip_df['end_station_name'].combine_first(trip_df['end station name'])
clean_trip_df['end_station_id']=trip_df['end_station_id'].combine_first(trip_df['end station id'])
clean_trip_df['start_lat']=trip_df['start_lat'].combine_first(trip_df['start station latitude'])
clean_trip_df['start_lng']=trip_df['start_lng'].combine_first(trip_df['start station longitude'])
clean_trip_df['end_lat']=trip_df['end_lat'].combine_first(trip_df['end station latitude'])
clean_trip_df['end_lng']=trip_df['end_lng'].combine_first(trip_df['end station longitude'])
clean_trip_df['member_casual']=trip_df['member_casual'].combine_first(trip_df['usertype'])

# Dropping the merged columns
clean_trip_df.drop(columns = ['starttime', 'stoptime', 'start station name', 'end station name',
                             'end station id', 'start station latitude', 'start station longitude',
                             'end station latitude', 'end station longitude', 'usertype',
                             'start station id'], inplace = True)

```
